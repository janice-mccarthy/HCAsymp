```r
ntrials<-1000

HC_stats<-NULL 

ts<-seq(0.001,.99999,by=.001)
ninterval <- length(ts)

W<-rep(0,ninterval)
U<-rep(0,ntrials)


for (n in c(100,1000,10000,50000)) {   #This is the asymptotic result - taking number of tests to very large 
   for (k in 1:ntrials) {  #Repeat for a number of generated datasets
    for (i in 1:ninterval) { #Get maximum over t for fixed number of tests
         t <- ts[i]
         S <- rbinom(1,n,prob=t)
         W[i] <- (S-n*t)/(sqrt(n*t*(1-t)))        
    }
    U[k] <- max(W)
 }
    HC_stats<-c(HC_stats,U)
#    hist(U,breaks=100,freq=FALSE)
}
```

```r
ntrials<-1000

K_stats<-NULL 

ts<-seq(0.001,.99999,by=.001)
ninterval <- length(ts)

W<-rep(0,ninterval)
U<-rep(0,ntrials)

par(mfrow=c(2,2))
for (n in c(100,1000,10000,50000)) {   #This is the asymptotic result - taking number of tests to very large 
   for (k in 1:ntrials) {  #Repeat for a number of generated datasets
    for (i in 1:ninterval) { #Get maximum over t for fixed number of tests
         t <- ts[i]
         S <- rbinom(1,n,prob=t)
         W[i] <- sqrt(n)*(S/n-t)        
    }
    U[k] <- max(W)
 }
    K_stats<-c(K_stats,U)
#    hist(U,breaks=100,freq=FALSE)
}
```
